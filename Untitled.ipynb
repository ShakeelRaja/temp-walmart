```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
plt.style.use('seaborn')
walmart = pd.read_csv('walmart_dataset.csv')
walmart.head()
```

```python
walmart.Weekly_Sales = (walmart.Weekly_Sales - walmart.Weekly_Sales.min()) / (walmart.Weekly_Sales.max() - walmart.Weekly_Sales.min())

walmart.Weekly_Sales.plot()
```

```python
for column in ['Store', 'Dept', 'Size',
       'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']:
    walmart[column].plot.hist(normed=True )
    walmart[column].plot.kde(label=column )
    plt.legend()
    plt.show()
```
